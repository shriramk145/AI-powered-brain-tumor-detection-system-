```python
import os
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow.keras.backend as K

# -------------------------------
# Paths & Config
# -------------------------------
IMAGE_DIR = "unet/image"
MASK_DIR = "unet/mask"

IMG_HEIGHT, IMG_WIDTH = 128, 128
NUM_CLASSES = 4 # Classes: 0=No Tumor, 1=Glioma, 2=Meningioma, 3=Pituitary

# -------------------------------
# Load Dataset
# -------------------------------
def load_data(image_dir, mask_dir, img_height, img_width):
    images, masks = [], []

    # Get a list of class folders from the image directory
    class_folders = sorted(os.listdir(image_dir))

    # Iterate through each class folder
    for class_id_str in class_folders:
        class_id = int(class_id_str)
        img_folder = os.path.join(image_dir, class_id_str)
        mask_folder = os.path.join(mask_dir, class_id_str)

        img_files = sorted(os.listdir(img_folder))
        mask_files = sorted(os.listdir(mask_folder))

        for img_file, mask_file in zip(img_files, mask_files):
            img_path = os.path.join(img_folder, img_file)
            mask_path = os.path.join(mask_folder, mask_file)

            # Load and preprocess image
            img = cv2.imread(img_path, cv2.IMREAD_COLOR)
            if img is None:
                print(f"Skipping image: {img_path}")
                continue
            img = cv2.resize(img, (img_width, img_height))
            images.append(img)

            # Load and preprocess mask
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is None:
                print(f"Skipping mask: {mask_path}")
                continue
            mask = cv2.resize(mask, (img_width, img_height), interpolation=cv2.INTER_NEAREST)

            # Assign correct class_id to the mask pixels
            # We assume your original mask files are binary (0 and non-zero)
            # and you want to classify non-zero pixels by class_id.
            mask[mask > 0] = class_id
            masks.append(mask)

    images = np.array(images, dtype=np.float32) / 255.0
    masks = np.array(masks, dtype=np.int32)

    # One-hot encode masks to create a separate channel for each class
    masks = tf.keras.utils.to_categorical(masks, num_classes=NUM_CLASSES)

    return images, masks

# -------------------------------
# Loss Functions
# -------------------------------
def dice_loss(y_true, y_pred, smooth=1e-6):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def combined_loss(y_true, y_pred):
    return tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred) + dice_loss(y_true, y_pred)

# -------------------------------
# U-Net Model
# -------------------------------
def unet_model(input_size=(IMG_HEIGHT, IMG_WIDTH, 3), num_classes=NUM_CLASSES):
    inputs = layers.Input(input_size)

    # Encoder
    c1 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(inputs)
    c1 = layers.BatchNormalization()(c1)
    c1 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(c1)
    c1 = layers.BatchNormalization()(c1)
    p1 = layers.MaxPooling2D((2,2))(c1)
    p1 = layers.Dropout(0.1)(p1)

    c2 = layers.Conv2D(128, (3,3), activation='relu', padding='same')(p1)
    c2 = layers.BatchNormalization()(c2)
    c2 = layers.Conv2D(128, (3,3), activation='relu', padding='same')(c2)
    c2 = layers.BatchNormalization()(c2)
    p2 = layers.MaxPooling2D((2,2))(c2)
    p2 = layers.Dropout(0.1)(p2)

    c3 = layers.Conv2D(256, (3,3), activation='relu', padding='same')(p2)
    c3 = layers.BatchNormalization()(c3)
    c3 = layers.Conv2D(256, (3,3), activation='relu', padding='same')(c3)
    c3 = layers.BatchNormalization()(c3)
    p3 = layers.MaxPooling2D((2,2))(c3)
    p3 = layers.Dropout(0.2)(p3)

    c4 = layers.Conv2D(512, (3,3), activation='relu', padding='same')(p3)
    c4 = layers.BatchNormalization()(c4)
    c4 = layers.Conv2D(512, (3,3), activation='relu', padding='same')(c4)
    c4 = layers.BatchNormalization()(c4)
    p4 = layers.MaxPooling2D((2,2))(c4)
    p4 = layers.Dropout(0.2)(p4)

    # Bottleneck
    c5 = layers.Conv2D(1024, (3,3), activation='relu', padding='same')(p4)
    c5 = layers.BatchNormalization()(c5)
    c5 = layers.Conv2D(1024, (3,3), activation='relu', padding='same')(c5)
    c5 = layers.BatchNormalization()(c5)
    c5 = layers.Dropout(0.3)(c5)

    # Decoder
    u6 = layers.Conv2DTranspose(512, (2,2), strides=(2,2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    c6 = layers.Conv2D(512, (3,3), activation='relu', padding='same')(u6)
    c6 = layers.BatchNormalization()(c6)
    c6 = layers.Conv2D(512, (3,3), activation='relu', padding='same')(c6)
    c6 = layers.BatchNormalization()(c6)

    u7 = layers.Conv2DTranspose(256, (2,2), strides=(2,2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    c7 = layers.Conv2D(256, (3,3), activation='relu', padding='same')(u7)
    c7 = layers.BatchNormalization()(c7)
    c7 = layers.Conv2D(256, (3,3), activation='relu', padding='same')(c7)
    c7 = layers.BatchNormalization()(c7)

    u8 = layers.Conv2DTranspose(128, (2,2), strides=(2,2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    c8 = layers.Conv2D(128, (3,3), activation='relu', padding='same')(u8)
    c8 = layers.BatchNormalization()(c8)
    c8 = layers.Conv2D(128, (3,3), activation='relu', padding='same')(c8)
    c8 = layers.BatchNormalization()(c8)

    u9 = layers.Conv2DTranspose(64, (2,2), strides=(2,2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    c9 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(u9)
    c9 = layers.BatchNormalization()(c9)
    c9 = layers.Conv2D(64, (3,3), activation='relu', padding='same')(c9)
    c9 = layers.BatchNormalization()(c9)

    outputs = layers.Conv2D(num_classes, (1,1), activation='softmax')(c9)

    model = models.Model(inputs=[inputs], outputs=[outputs])
    return model

# -------------------------------
# Main Execution
# -------------------------------
if __name__ == "__main__":
    # Load Dataset
    images, masks = load_data(IMAGE_DIR, MASK_DIR, IMG_HEIGHT, IMG_WIDTH)
    print("Dataset loaded:", images.shape, masks.shape)

    # Train-test split
    X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

    # Build and compile model
    model = unet_model()
    model.compile(optimizer='adam', loss=combined_loss, metrics=['accuracy'])
    model.summary()

    # Train Model
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=7,
        batch_size=16,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
            tf.keras.callbacks.ModelCheckpoint("best_model.h5", save_best_only=True)
        ]
    )

    # Save Model
    #model.save("unet_brain_tumor.h5")
    #print("Model saved as unet_brain_tumor.h5")

    # -------------------------------
    # Prediction & Visualization
    # -------------------------------
    def display_sample(X, y, preds, index=0):
        """Displays the original image, true mask, and predicted mask."""
        plt.figure(figsize=(15,5))

        plt.subplot(1,3,1)
        plt.title("Image")
        plt.imshow(X[index])

        plt.subplot(1,3,2)
        plt.title("True Mask")
        plt.imshow(np.argmax(y[index], axis=-1), cmap="jet")

        plt.subplot(1,3,3)
        plt.title("Predicted Mask")
        plt.imshow(np.argmax(preds[index], axis=-1), cmap="jet")

        plt.show()

    preds = model.predict(X_val)

    # You can change the index to view different examples
    display_sample(X_val, y_val, preds, index=15)
```

```python
import os
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt

# --- Same configurations as in the training script ---
IMG_HEIGHT, IMG_WIDTH = 128, 128

# --- Define custom objects for loading the model ---
# The model needs to know about the custom loss function when you load it.
def dice_loss(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def combined_loss(y_true, y_pred):
    return tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred) + dice_loss(y_true, y_pred)

# -------------------------------
# Load Model and Predict
# -------------------------------

def predict_on_image(image_path, saved_model_path="unet_brain_tumor.h5"):
    """
    Loads a saved U-Net model and performs segmentation on a single image.
    :param image_path: Path to the input image.
    :param saved_model_path: Path to the saved model file.
    """

    # Check if the model file exists
    if not os.path.exists(saved_model_path):
        print(f"Error: Model file not found at {saved_model_path}")
        return

    # Load the trained model with custom objects
    # This is crucial for models with custom loss functions
    with tf.keras.utils.custom_object_scope({'combined_loss': combined_loss}):
        model = tf.keras.models.load_model(saved_model_path)

    # Preprocess the new image
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if img is None:
        print(f"Error: Could not read image from {image_path}")
        return

    img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    # The model expects a batch of images, so we add a new axis
    img_normalized = np.expand_dims(img, axis=0) / 255.0

    # Make prediction
    prediction = model.predict(img_normalized, verbose=0)

    # The prediction is a one-hot encoded tensor; convert it back to a single mask
    predicted_mask = np.argmax(prediction[0], axis=-1)

    # Display the results
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 2, 2)
    plt.title("Predicted Mask")
    plt.imshow(predicted_mask, cmap="jet")

    plt.show()

# -------------------------------
# Example Usage
# -------------------------------
if __name__ == "__main__":
    # --- CHANGE THIS PATH TO YOUR IMAGE ---
    # Make sure you provide a valid path to an image file.
    test_image_path = r"D:\PycharmProject\AI & DL\hackathon\brain\Testing\meningioma\Te-me_0020.jpg"

    predict_on_image(test_image_path)
```

```python
# Load model back
from tensorflow.keras.models import load_model

model = load_model("unet_brain_tumor.h5", compile=False)

def predict_single_image(image_path, model, img_height=128, img_width=128):
    # Load image
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    if img is None:
        raise ValueError(f"Image not found: {image_path}")

    original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # keep copy for visualization
    img_resized = cv2.resize(img, (img_width, img_height))
    img_resized = img_resized.astype(np.float32) / 255.0
    img_resized = np.expand_dims(img_resized, axis=0)  # add batch dim

    # Predict
    pred = model.predict(img_resized)
    pred_mask = np.argmax(pred[0], axis=-1)

    # Resize mask back to original image size
    pred_mask = cv2.resize(pred_mask.astype(np.uint8), (original.shape[1], original.shape[0]), interpolation=cv2.INTER_NEAREST)

    # Display
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.title("Original Image")
    plt.imshow(original)

    plt.subplot(1,2,2)
    plt.title("Predicted Mask")
    plt.imshow(pred_mask, cmap="jet")
    plt.colorbar()
    plt.show()

    return pred_mask


# Example: run prediction on new image
mask = predict_single_image(r"D:\PycharmProject\AI & DL\hackathon\brain\Training\pituitary\Tr-pi_0197.jpg", model)
```
